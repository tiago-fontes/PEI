```python
import pandas as pd
```

```python
data = pd.read_csv('raw_data_com_lixo.csv')
```

```python
data[(data['sensors.pm25']<=90) & (data['sensors.classification']==1)]
data[(data['sensors.pm25']>=40) & (data['sensors.classification']==0)]
data[(data['sensors.pm25']<=0.8)]
data = data.drop(['Unnamed: 0','sensors.classification'],axis=1)
data = data.drop(['sensors.temperature','sensors.humidity','sensors.pressure'],axis=1)
data = data.drop(['sensors.gas'],axis=1)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=111)
kmeans.fit(data)
```

```python
import matplotlib.pyplot as plt
```

```python
for c1,col1 in enumerate(data.columns):
    for c2,col2 in enumerate(data.columns):
        if(c1 >= c2):
            ;
        else:
            plt.scatter(data[kmeans.labels_==0][col1], data[kmeans.labels_==0][col2], s=100, c='red', label ='Cluster 0')
            plt.scatter(data[kmeans.labels_==1][col1], data[kmeans.labels_==1][col2], s=100, c='green', label ='Cluster 1')
            plt.scatter(kmeans.cluster_centers_[:, c1], kmeans.cluster_centers_[:, c2], s=50, c='yellow', label = 'Centroids')
            plt.title('Clusters ' + str(col1) + ' x ' + str(col2))
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.show()
```
